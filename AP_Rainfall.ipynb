```python
import os

files = [os.path.join(path, f) for f in os.listdir(path) if f.endswith('.csv')]

print(files)
```

```python
import pandas as pd

df_list = []

for file in files:
    temp = pd.read_csv(file, nrows=50000)
    temp.columns = temp.columns.str.strip().str.lower()
    df_list.append(temp)

print("Files loaded:", len(df_list))
```

```python
final_df = pd.concat(df_list, ignore_index=True)

print("Shape:", final_df.shape)
final_df.head()
```

```python
print(files)
```

```python
final_df = final_df.sample(50000)

print("New Shape:", final_df.shape)
```

```python
final_df['date'] = pd.to_datetime(final_df['date'])

final_df['year'] = final_df['date'].dt.year
final_df['month'] = final_df['date'].dt.month
final_df['day'] = final_df['date'].dt.day
```

```python
final_df = final_df.sort_values(by=['village', 'date'])
```

```python
final_df = final_df[final_df['rainfall_mm'] > 0]
```

```python
final_df['rain_lag1'] = final_df.groupby('village')['rainfall_mm'].shift(1)
final_df['rain_lag2'] = final_df.groupby('village')['rainfall_mm'].shift(2)

final_df = final_df.dropna()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

features = ['year','month','day','centroid_lat','centroid_lon','rain_lag1','rain_lag2']
target = 'rainfall_mm'

X = final_df[features]
y = final_df[target]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

model = RandomForestRegressor(n_estimators=5)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error

pred = model.predict(X_test)

print("MAE:", mean_absolute_error(y_test, pred))
print("Pred:", pred[:5])
print("Actual:", y_test[:5].values)
```

```python
def get_risk(rain):
    if rain > 50:
        return "HIGH"
    elif rain > 20:
        return "MEDIUM"
    else:
        return "LOW"
```

```python
final_df['predicted_rain'] = model.predict(X)
final_df['risk_level'] = final_df['predicted_rain'].apply(get_risk)

final_df[['village','predicted_rain','risk_level']].head()
```

```python
def predict_with_risk(year, month, day, lat, lon, lag1, lag2):
    rain = model.predict([[year, month, day, lat, lon, lag1, lag2]])[0]
    
    if rain > 20:
        risk = "HIGH"
    elif rain > 50:
        risk = "MEDIUM"
    else:
        risk = "LOW"
    
    return rain, risk
import warnings
warnings.filterwarnings('ignore')
```

```python
def predict_for_village(village_name, year, month, day):
    
    # get village data
    data = final_df[final_df['village'] == village_name].iloc[-1]
    
    lat = data['centroid_lat']
    lon = data['centroid_lon']
    lag1 = data['rain_lag1']
    lag2 = data['rain_lag2']
    
    # prediction
    rain = model.predict([[year, month, day, lat, lon, lag1, lag2]])[0]
    
    # risk
    if rain > 20:
        risk = "HIGH"
    elif rain > 5:
        risk = "MEDIUM"
    else:
        risk = "LOW"
    
    print("Village:", village_name)
    print("Predicted Rainfall:", rain)
    print("Risk Level:", risk)
    
    return rain, risk, lat, lon
```

```python
def predict_all(year, month, day):
    
    df_copy = final_df.copy()
    
    # Use same input date for all rows
    df_copy['year'] = year
    df_copy['month'] = month
    df_copy['day'] = day
    
    features = ['year','month','day','centroid_lat','centroid_lon','rain_lag1','rain_lag2']
    
    # Predict
    df_copy['predicted_rain'] = model.predict(df_copy[features])
    
    # Risk logic
    def get_risk(rain):
        if rain > 20:
            return "HIGH"
        elif rain > 5:
            return "MEDIUM"
        else:
            return "LOW"
    
    df_copy['risk_level'] = df_copy['predicted_rain'].apply(get_risk)
    
    # Final table
    result = df_copy[['district','mandal','village','predicted_rain','risk_level']]
    
    return result
```
